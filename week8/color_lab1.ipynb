```python
y1= 163
y2= 258
x1= 206
x2= 233
```

```python
''' 
this patch of image contains the yellow pixels, 
that patch of a yellow pokemon was found manualy. we use it to: 
1- in the Lab space
2- find the min and max of a and b of that patch  
3- apply a color filter and keep the pixels between that min and max and delete the other pixels

'''
```

```python
# showing that manualy found and set yellow patch
img3 = img[x1:x2 , y1:y2, :]
plt.imshow(cv2.cvtColor( img3 , cv2.COLOR_BGR2RGB))
```

```python
# finding the min and max A and b of that patch
# then we can use them in our color filter
#
amin = np.min(labp[x1:x2 , y1:y2, 1])
amax = np.max(labp[x1:x2 , y1:y2, 1])
bmin = np.min(labp[x1:x2 , y1:y2, 2])
bmax = np.max(labp[x1:x2 , y1:y2, 2])
```

```python
# showing the mins and maxs
amin, amax, bmin, bmax
```

```python
# color filter: developing a mask of [0,1]. 0 when the pixel's a or b are outside the a,b range
# of the yellow pokemon, while it is 1 when pixel's a,b are inside the range
#
M = labp.shape
lab_mask = np.zeros( M ) 
for i in range(M[0]): 
    for j in range(M[1]): 
        if ( amin < labp[i,j,1] < amax) and (bmin < labp[i,j,2] < bmax):
            lab_mask[i,j,:] = 1
```

```python
# inverse transform from Lab to BGR
# and pixelwise multiplication of the mask with the inverted-to-BGR image
#
img3 = cv2.cvtColor( labp , cv2.COLOR_Lab2BGR)
img4 = np.uint8( np.multiply( img3 , lab_mask) )
```

```python
# subplot of the results of color filtering
#
plt.subplots(2,2,figsize=(12,12))
plt.subplot(221),plt.imshow(cv2.cvtColor( img3 , cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow( np.uint8( lab_mask *250) , cmap = 'gray')
plt.title('Mask'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow( cv2.cvtColor( img4 , cv2.COLOR_BGR2RGB) , cmap = 'gray')
plt.title('masked image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# showing the results of color filtering, mask and resulted image
#
cv2.imshow("The Masked Image" , img4)
cv2.imshow("The Mask" , np.uint8( lab_mask *250))
```
