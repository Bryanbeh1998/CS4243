```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Load the image IMG_0720.jpg in set1

```python
im_path = "../../Examples/set1/IMG_0720.jpg"
im = cv2.imread(im_path)
```

### Display the image

```python
cv2.namedWindow("Original", cv2.WINDOW_NORMAL)
cv2.imshow("Original", im)
cv2.waitKey()
```

### Check the size, dimension, type, min and max value of the image

```python
print(f"The size of the image is: {im.size}")
print(f"The dimension of the image is: {im.shape}")
print(f"The type of the image is: {type(im)}")
print(f"The data type of the image is: {im.dtype}")
print(f"The max value of the image is: {im.max()}")
print(f"The min value of the image is: {im.min()}")
```

### Visualize the first 600 columns of the colored image

```python
cv2.namedWindow("First 600 columns", cv2.WINDOW_NORMAL)
cv2.imshow("First 600 columns", im[:,:600,:])
cv2.waitKey()
```

### Visualize the first 600 rows of the colored image

```python
cv2.namedWindow("First 600 rows", cv2.WINDOW_NORMAL)
cv2.imshow("First 600 rows", im[:600, :, :])
cv2.waitKey()
```

### Visualize the R, G, B channel of the image separately, does the visualization show the channels as you expected?

```python
# CV2 read the image as BGR instead of RGB
cv2.namedWindow("Red", cv2.WINDOW_NORMAL)
cv2.imshow("Red", im[:, :, -1])
cv2.waitKey()
cv2.namedWindow("Green", cv2.WINDOW_NORMAL)
cv2.imshow("Green", im[:, :, 1])
cv2.waitKey()
cv2.namedWindow("Blue", cv2.WINDOW_NORMAL)
cv2.imshow("Blue", im[:, :, 0])
cv2.waitKey()
```

```python
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.rc('figure', figsize=(20, 8))
plt.imshow(np.hstack((im, im_rgb)))
```

### Convert the image to grayscale, then visualize the output

```python
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
cv2.namedWindow("Grayscale", cv2.WINDOW_NORMAL)
cv2.imshow("Grayscale", im_gray)
cv2.waitKey()
```

### Create negtive images for colored image and the grayscale image, then visualize the output

```python
im_neg = 255-im
cv2.namedWindow("negative_image", cv2.WINDOW_NORMAL)
cv2.imshow("negative_image", im_neg)
cv2.waitKey()
```

```python
im_gray_neg = 255-im_gray
cv2.namedWindow("negative_gray_image", cv2.WINDOW_NORMAL)
cv2.imshow("negative_gray_image", im_gray_neg)
cv2.waitKey()
```

### Save images into png format, negative_image.png,  negative_gray_image.png

```python
cv2.imwrite("negative_image.png", im_neg)
cv2.imwrite("negative_gray_image.png", im_gray_neg) 
```

### Rotate the negative image by 90, 180 and 270 degree clockwise and visualize the results

```python
im_neg_90 = cv2.rotate(im_neg, cv2.ROTATE_90_CLOCKWISE)
im_neg_180 = cv2.rotate(im_neg, cv2.ROTATE_180)
im_neg_270 = cv2.rotate(im_neg, cv2.ROTATE_90_COUNTERCLOCKWISE)

cv2.namedWindow("90", cv2.WINDOW_NORMAL)
cv2.imshow("90", im_neg_90)
cv2.waitKey()
cv2.namedWindow("180", cv2.WINDOW_NORMAL)
cv2.imshow("180", im_neg_180)
cv2.waitKey()
cv2.namedWindow("270", cv2.WINDOW_NORMAL)
cv2.imshow("270", im_neg_270)
cv2.waitKey()
```

### Resize the original image to [200, 300] pixels, and [4000, 6000] pixels. What are the options?

```python
im_small = cv2.resize(im, (200,300))
im_large = cv2.resize(im, dsize=(4000, 2000))
im_large_cube = cv2.resize(im, dsize=(4000, 2000), interpolation=cv2.INTER_AREA)
(im_large-im_large_cube).max()
```

```python
cv2.namedWindow("small", cv2.WINDOW_NORMAL)
cv2.imshow("small", im_small)
cv2.waitKey()
cv2.namedWindow("large", cv2.WINDOW_NORMAL)
cv2.imshow("large", im_large)
cv2.waitKey()
cv2.namedWindow("large cube", cv2.WINDOW_NORMAL)
cv2.imshow("large cube", im_large_cube)
cv2.waitKey()
cv2.namedWindow("large diff", cv2.WINDOW_NORMAL)
cv2.imshow("diff", im_large-im_large_cube)
cv2.waitKey()
```

```python
cv2.destroyAllWindows()
for i in range(10):
    cv2.waitKey(1)
```
